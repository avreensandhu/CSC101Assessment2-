```python
#STEP5 Simplified implementation:
def is_leap_year(year):
    """
    Determine if a year is a leap year using the three sequential rules,
    making the order of operations explicit with if/elif/else.
    """
    # 1. Highest Priority Rule: If divisible by 400, it IS a leap year.
    if year % 400 == 0:
        return True

    # 2. Middle Priority Rule: If divisible by 100 but NOT 400, it is NOT a leap year.
    #    (This check only happens if the 400 check failed above)
    elif year % 100 == 0:
        return False

    # 3. Lowest Priority Rule: If divisible by 4 but NOT 100 or 400, it IS a leap year.
    #    (This check only happens if the 400 and 100 checks failed above)
    elif year % 4 == 0:
        return True

    # 4. Default: If none of the above conditions were met, it is NOT a leap year.
    else:
        return False
```
