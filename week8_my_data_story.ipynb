```python
import os

file_path = '/content/my_data.txt'

# Check if the file exists
if not os.path.exists(file_path):
    print(f"Error: File not found at {file_path}")
else:
    with open(file_path, 'r') as f:
        content = f.read()

    # Number of lines
    lines = content.strip().split('\n')
    num_lines = len(lines)

    # Number of words
    words = content.split()
    num_words = len(words)

    # Longest line
    longest_line = max(lines, key=len)

    # First and last lines
    first_line = lines[0] if num_lines > 0 else ""
    last_line = lines[-1] if num_lines > 0 else ""

    # Total character count
    total_characters = len(content)

    print(f"--- File Analysis for '{file_path}' ---")
    print(f"Number of lines: {num_lines}")
    print(f"Number of words: {num_words}")
    print(f"Longest line: '{longest_line}' (length: {len(longest_line)} characters)")
    print(f"First line: '{first_line}'")
    print(f"Last line: '{last_line}'")
    print(f"Total character count: {total_characters}")
    print("-----------------------------------------")
```
