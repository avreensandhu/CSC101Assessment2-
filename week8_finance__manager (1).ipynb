## Define Parse Transaction Function

### Subtask:
Create the `parse_transaction(line)` function. This function will take a single transaction string, attempt to split it into description and amount, convert the amount to a float, and return a dictionary {'description': str, 'amount': float}. It will include `try-except` blocks to handle empty lines, lines without a comma (IndexError), and lines with non-numeric amounts (ValueError), returning `None` for any corrupted lines.


**Reasoning**:
Define the `parse_transaction` function as specified in the instructions, including `try-except` blocks to handle various parsing errors and return `None` for corrupted lines.



```python
def parse_transaction(line):
    """Parses a single transaction string into a dictionary.

    Args:
        line (str): The transaction string in 'description, amount' format.

    Returns:
        dict: A dictionary {'description': str, 'amount': float} if parsing is successful.
        None: If the line is empty, malformed, or contains non-numeric amounts.
    """
    try:
        if not line or line.isspace():
            return None

        parts = line.split(',', 1)
        if len(parts) != 2:
            # This handles cases where there's no comma, or too many (though split(..., 1) limits this)
            raise IndexError("Line does not contain a comma separating description and amount.")

        description = parts[0].strip()
        amount_str = parts[1].strip()
        amount = float(amount_str)

        return {'description': description, 'amount': amount}
    except ValueError:
        print(f"Skipping line '{line}': Amount is not a valid number.")
        return None
    except IndexError:
        print(f"Skipping line '{line}': Incorrect format (missing comma).")
        return None
    except Exception as e:
        print(f"Skipping line '{line}': An unexpected error occurred: {e}")
        return None

print("The 'parse_transaction' function has been defined.")
```

## Define Process Transactions Function

### Subtask:
Create a `process_transactions(raw_lines)` function. This function will iterate through a list of raw transaction strings. For each line, it will call `parse_transaction`. It will collect all successfully parsed transaction dictionaries into a list and maintain a counter for how many lines failed to parse. It should return both the list of parsed transactions and the count of failed lines.


**Reasoning**:
Define the `process_transactions` function as specified, which iterates through raw transaction lines, uses `parse_transaction`, and aggregates successful parses while counting failures.



```python
def process_transactions(raw_lines):
    """Processes a list of raw transaction strings.

    Args:
        raw_lines (list): A list of raw transaction strings.

    Returns:
        tuple: A tuple containing:
            - list: A list of successfully parsed transaction dictionaries.
            - int: The count of lines that failed to parse.
    """
    parsed_transactions = []
    failed_count = 0

    for line in raw_lines:
        parsed = parse_transaction(line)
        if parsed is not None:
            parsed_transactions.append(parsed)
        else:
            failed_count += 1

    return parsed_transactions, failed_count

print("The 'process_transactions' function has been defined.")
```

## Define Analyze Transactions Function

### Subtask:
Create an `analyze_transactions(parsed_transactions)` function. This function will take a list of successfully parsed transaction dictionaries. It will calculate the total number of valid transactions, total income (sum of positive amounts), total expenses (sum of negative amounts), and the net amount (income + expenses). It should return these four calculated values.


**Reasoning**:
To implement the `analyze_transactions` function, I will follow the instructions to initialize variables, iterate through the parsed transactions, categorize amounts into income and expenses, and calculate the net amount.



```python
def analyze_transactions(parsed_transactions):
    """Analyzes a list of parsed transactions to calculate income, expenses, and net amount.

    Args:
        parsed_transactions (list): A list of dictionaries, each with 'description' and 'amount' keys.

    Returns:
        tuple: (total_transactions, total_income, total_expenses, net_amount)
    """
    total_income = 0.0
    total_expenses = 0.0
    total_transactions = 0

    for transaction in parsed_transactions:
        total_transactions += 1
        amount = transaction['amount']
        if amount > 0:
            total_income += amount
        else:
            total_expenses += amount # Expenses are typically negative or zero amounts

    net_amount = total_income + total_expenses

    return total_transactions, total_income, total_expenses, net_amount

print("The 'analyze_transactions' function has been defined.")
```
