```python
bill = 45.80
tip_percentage = 0.15

tip_amount = bill * tip_percentage
total_amount = bill + tip_amount

print(f"Bill: ${bill:.2f}")
print(f"Tip Percentage: {tip_percentage * 100:.0f}%")
print(f"Tip Amount: ${tip_amount:.2f}")
print(f"Total Amount: ${total_amount:.2f}")
```
